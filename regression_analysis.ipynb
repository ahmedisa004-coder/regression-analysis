# Simple Linear Regression - Marketing ROI Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv('marketing_data.csv')
print(df.head())
print(df.shape)
print(df.isnull().sum())
```

## EDA - TV has highest correlation with Sales

```python
print(df.corr())
plt.scatter(df['TV'], df['Sales'])
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('TV vs Sales')
plt.show()
```

## OLS Regression Model

```python
X = sm.add_constant(df['TV'])
y = df['Sales']
model = sm.OLS(y, X).fit()
print(model.summary())
```

## Diagnostic Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))
axes[0].scatter(model.fittedvalues, model.resid)
axes[0].axhline(0, color='red')
axes[0].set_title('Residuals vs Fitted')
sm.qqplot(model.resid, line='s', ax=axes[1])
axes[1].set_title('Q-Q Plot')
plt.tight_layout()
plt.show()
```

## Conclusion

TV advertising has the strongest correlation with Sales (highest R-squared).

**Recommendation:** Allocate the largest portion of the marketing budget to TV advertising as it delivers the strongest ROI impact on Sales.
